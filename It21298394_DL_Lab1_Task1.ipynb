```python
arr2 = np.random.exponential(1,100000)
```

```python
plt.hist(arr2, density=True, bins=200, histtype="step", color="red", label="rand")
plt.axis([-1, 8, 0, 1.1])
plt.legend(loc = "upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection="3d")

num1 = np.arange(-5, 5, 0.1)
num2 = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(num1, num2)

Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="viridis", edgecolor="none")
ax.set_title("Surface plot")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True,
           hue='Stage')
```

```python
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
plt.figure(figsize=(9,6))
sns.boxplot(data=stats_df)
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Type 1', y='Attack', data=df)
```
